## Exercise 2: Zeros

```python
import numpy as np

# Create a 1D array of 300 zeros
zeros_array = np.zeros(300)

print(zeros_array)
```

```python
# Reshape the array to 3 rows and 100 columns
reshaped_array = zeros_array.reshape((3, 100))

print(reshaped_array)
# print(f"Shape: {reshaped_array.shape}")
```
